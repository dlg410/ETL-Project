```python
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

from time import sleep
from random import randint
```

```python
# https://splinter.readthedocs.io/en/latest/drivers/chrome.html
!which chromedriver
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'http://hdr.undp.org/en/countries'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
articles = soup.find_all('div', class_="countriesListItem")
```

```python
for article in articles:
    print ('http://hdr.undp.org/en/' + article.find('a')['href'])
    print (article.find('a')['href'])
    print(article.find('a').text)
```

```python
links= []
countries= []

for article in articles:
    country = (article.find('a').text)
    link = ('http://hdr.undp.org/en/' + article.find('a')['href'])
#     print (article.find('a')['href'])
#     print(article.find('a').text)
    countries.append(country)
    links.append(link)
```

```python
UN_table = ({
    'Country': countries,
    'Link': links,
})
UN_table
```
