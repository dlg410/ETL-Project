```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine 
```

```python
engine = create_engine('postgresql://postgres:password@localhost/HDI vs WHR')
conn = engine.connect()
```

```python
data_df=pd.read_csv("../../1-Extract/Raw data/Scraped Data/Human Development Index.csv")
```

```python
data_df
```

```python
data_df.drop(data_df.columns[data_df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
data_df
```

```python
clean_df = (data_df.drop(data_df.iloc[:, 2:22], axis = 1).dropna(0))
clean_df
```

```python
del clean_df['HDI Rank']
clean_df
```

```python
clean_df.to_csv("../Cleaned/Human Development Index.csv")
```

```python
clean_df.to_sql(
    name='Human Development Index',
    con=engine,
    index=False
)
```
